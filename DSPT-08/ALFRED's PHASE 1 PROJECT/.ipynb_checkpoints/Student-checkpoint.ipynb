### Electric Vehicle Dataset: Overview

This dataset provides information on Battery Electric Vehicles (BEVs) and Plug-in Hybrid Electric Vehicles (PHEVs) that are currently registered through the Washington State Department of Licensing (DOL). The dataset is sourced from the US government and is shared on Kaggle.

#### Dataset Source: Kaggle

##### Dataset Description:  
- This dataset contains detailed records of electric vehicles registered in Washington State, specifically focusing on Battery Electric Vehicles (BEVs) and Plug-in Hybrid Electric Vehicles (PHEVs).
- It captures various aspects of these vehicles, including their registration details, technical specifications, and other relevant attributes that provide insights into the electric vehicle market.

##### Data Source:  
- The data is sourced from the Washington State Department of Licensing (DOL) and is made publicly available by the US government.

##### Features of the Dataset:
1. Vehicle Type: Identifies whether the vehicle is a Battery Electric Vehicle (BEV) or a Plug-in Hybrid Electric Vehicle (PHEV).
2. Brand: The manufacturer or brand of the electric vehicle.
3. Model: The specific model of the electric vehicle.
4. Battery Capacity: The capacity of the vehicle's battery, typically measured in kilowatt-hours (kWh).
5. Sales: The number of units sold or registered, indicating the popularity and market penetration of the vehicle.
6. Cost: The cost associated with the vehicle, which could include the purchase price or other financial metrics.
7. Region: The geographical region within Washington State where the vehicle is registered, providing insights into regional adoption rates.
8. Incentives: Information on any financial incentives or subsidies applied to the vehicle, which can impact sales and profitability.
