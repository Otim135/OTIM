### Feature Engineering

#### 1. Extracting and Converting Dates
  * Objective: Create features related to the timing of incidents, such as the year, month, day of the week, and whether it was a weekend or holiday.

```python
# Ensure 'Event.Date' is in datetime format
aviation_data['Event.Date'] = pd.to_datetime(aviation_data['Event.Date'], errors='coerce')

# Extract year, month, and day of week from the 'Event.Date'
aviation_data['Event.Year'] = aviation_data['Event.Date'].dt.year
aviation_data['Event.Month'] = aviation_data['Event.Date'].dt.month
aviation_data['Event.DayOfWeek'] = aviation_data['Event.Date'].dt.dayofweek

# Determine if the incident occurred on a weekend
aviation_data['Event.IsWeekend'] = aviation_data['Event.DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)
```

#### 2. Categorizing Aircraft Models
  * Objective: Group similar aircraft models into broader categories to reduce complexity and improve analysis.

```python
# Create broader categories for aircraft models (this is just a sample grouping, and should be adjusted based on the dataset specifics)
def categorize_model(model):
    if 'Boeing' in model:
        return 'Boeing'
    elif 'Airbus' in model:
        return 'Airbus'
    elif 'Cessna' in model:
        return 'Cessna'
    elif 'Piper' in model:
        return 'Piper'
    else:
        return 'Other'

aviation_data['Aircraft.Model.Category'] = aviation_data['Model'].apply(categorize_model)
```

#### 3. Engineering Total Injuries
  * Objective: Create a comprehensive feature that aggregates different types of injuries.

```python
# Calculate total injuries by summing up different types of injuries
aviation_data['Total.Injuries'] = aviation_data[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries']].sum(axis=1)
```

#### 4. Weather Conditions Categorization
  * Objective: Simplify weather conditions into broader categories (e.g., Visual, Instrument, Unknown) for better analysis.

```python
# Categorize weather conditions into broader categories
def categorize_weather(condition):
    if condition == 'VMC':
        return 'Visual'
    elif condition == 'IMC':
        return 'Instrument'
    else:
        return 'Unknown'

aviation_data['Weather.Condition.Category'] = aviation_data['Weather.Condition'].apply(categorize_weather)
```

#### 5. Feature for Severity of Incidents
  * Objective: Create a severity index based on the number and type of injuries to quantify the severity of each incident.

```python
# Define a severity index based on injuries
def severity_index(row):
    if row['Total.Fatal.Injuries'] > 0:
        return 'Fatal'
    elif row['Total.Serious.Injuries'] > 0:
        return 'Serious'
    elif row['Total.Minor.Injuries'] > 0:
        return 'Minor'
    else:
        return 'No Injuries'

aviation_data['Incident.Severity'] = aviation_data.apply(severity_index, axis=1)
```

```python
# Encode flight phase into numeric categories
flight_phase_mapping = {'Takeoff': 1, 'Climb': 2, 'Cruise': 3, 'Descent': 4, 'Approach': 5, 'Landing': 6, 'Maneuvering': 7, 'Taxi': 8, 'Standing': 9, 'Unknown': 0}
aviation_data['Flight.Phase.Code'] = aviation_data['Broad.phase.of.flight'].map(flight_phase_mapping).fillna(0).astype(int)
```

#### 9. Time Series Plot of Incidents by Flight Phase Over Time
  * This plot shows how incidents in different flight phases vary over time, which can help identify trends.

```python
# Group data by year and flight phase to count the number of incidents
incidents_by_year_and_phase = aviation_data.groupby(['Event.Year', 'Broad.phase.of.flight']).size().reset_index(name='Incidents')

# Pivot the table for easier plotting
incidents_pivot = incidents_by_year_and_phase.pivot(index='Event.Year', columns='Broad.phase.of.flight', values='Incidents').fillna(0)

# Plot the data
plt.figure(figsize=(14, 8))
incidents_pivot.plot(ax=plt.gca(), marker='o')
plt.title('Incidents by Flight Phase Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend(title='Flight Phase')
plt.grid(True)
plt.show()
```
